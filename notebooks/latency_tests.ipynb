```python
from langchain_openai import AzureOpenAIEmbeddings

embeddings = AzureOpenAIEmbeddings(
        azure_endpoint=AZURE_AI_ENDPOINT,
        api_key=AZURE_AI_API_KEY,
        api_version="2024-02-15-preview",
        azure_deployment="text-embedding-ada-002"
    )
```

```python
vec1 = embeddings.embed_query("waterproof")
vec2 = embeddings.embed_query("show me lights which can work in rain")
```

```python
import numpy as np

def cosine_similarity_numpy(vec_a, vec_b):
    # Convert lists to NumPy arrays
    a = np.array(vec_a)
    b = np.array(vec_b)
    
    # Compute dot product and norms using matrix operations
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    
    if norm_a == 0 or norm_b == 0:
        return 0.0
        
    return dot_product / (norm_a * norm_b)
```

```python
cosine_similarity_numpy(vec1, vec2)
```
